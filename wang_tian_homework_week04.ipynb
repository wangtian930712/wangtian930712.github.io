```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#import ipywidgets
import ipywidgets
```

```python
#read the csv file
ufos = pd.read_csv("/Users/reiko/Downloads/ufo-scrubbed-geocoded-time-standardized-00.csv",
                  names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
```

```python
ufos = ufos.fillna({'country':'unknown', 'state': 'unknown', 'shape':'unknown'})
```

```python
#select a sample of the dataset
nSamples = 1000
np.random.seed(0)
randomIndicies = np.random.choice(range(len(ufos)-1), nSamples, replace=False)

ufosDS = ufos.loc[randomIndicies]
ufosDS
```

### Question 1

```python
#create two dropdown lists
#using codes from the official documents: https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20List.html
```

```python
#pick up x field

xvalue = ipywidgets.Dropdown(options=["date", "city", "state", "country",\
                                      "shape", "duration_seconds", \
                                      "report_date", "latitude", "longitude"],\
                            value='date', description='X field: ');
xvalue
```

```python
#pick up y field

yvalue = ipywidgets.Dropdown(options=["date", "city", "state", "country",\
                                      "shape", "duration_seconds", \
                                      "report_date", "latitude", "longitude"],\
                            value='date', description='Y field: ');
yvalue
```

```python
#create the scatter plot
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(ufosDS[xvalue.value], ufosDS[yvalue.value],'.')
ax.set_xlabel(xvalue.value)
ax.set_ylabel(yvalue.value)
plt.show()
```

```python
#pick a colormap

@ipywidgets.interact(colormap=plt.colormaps())
def plot_ufo(colormap):
    plt.figure(figsize=(20,10))
    plt.scatter(ufosDS[xvalue.value], ufosDS[yvalue.value], \
                c=np.log10(ufosDS['duration_seconds']), cmap=colormap)
```

Summary on Question 1: 

In Question 1, a scatter plot from the UFO dataset is designed that allows the user to change the x and y fields and to pick a colormap. For the x and y fields on a scatter plot, not all the columns in the dataset are provided to the user. Only "date", "city", "state", "country", "shape", "duration_seconds", "report_date", "latitude", and "longitude" (9 columns) are provided in this case while "comment" and "duration" are removed from the dropdown list. 
"comment" is not selected because the value for each comment in the dataset is very unique, and it is very difficult to put them into several categories. In other words, if the user selects "comment" for x or y field, the number of values in x-axis or y-axis will be almost equal to the number of instances in the dataset, which makes the visualization less effective and hard to read.
"duration" is not selected because it is a categorical variable in the dataset, so it has similar reason as explained for "comment". In addition, since the dataset includes a column called "duration_seconds" that indicates duration in seconds, we could use this variable as a replacement of "duration". 
Thus, all the columns except "comment" and "duration" are selected for this question. 

### Question 2

```python
#select a column

var = ipywidgets.Dropdown(options=["date", "city", "state", "country",\
                                      "shape", "duration_seconds", \
                                      "report_date", "latitude", "longitude"],\
                            value='date', description='Select a column: ', style = {'description_width': 'initial'});
var
```

```python
#select the number of bins if it is numerical

if var.value in ["duration_seconds", "latitude", "longitude"]:
    irange = ipywidgets.IntSlider(min = 0, max = 20, step = 1)
else:
    irange = None
irange
```

```python
#display binned, aggregate values of selected column

if var.value in ["duration_seconds", "latitude", "longitude"]:
    plt.clf()
    fig, ax = plt.subplots(figsize=(20,10))
    ufosDS.hist(column=var.value, ax=ax, bins=irange.value)
    plt.show()
else:
    ax = ufosDS[var.value].value_counts().plot(kind='bar', figsize=(15, 8))
    ax.set_xlabel(var.value)
```

Summary on Question 2

For the second question, the goal is to build a  widget that displays binned, aggregate values of a column selected by the user. First I created a dropdown list to let the user selects the column. If the column is numerical, it will follows by a slider to allow the user to select the number of bins. For Question 2, I use similar criteria from Question 1, where the column "comment" and "duration" are removed from the list. These two columns are removed because they are categorical variables that are very difficult to be binned or aggregated (especially for "comment" column, each instance in the dataset may have a unique comment). Therefore, all the columns except comment and duration are included in the dropdown list.
