```python
import bqplot
from bqplot import pyplot as plt
import ipywidgets
```

```python
def create_scatter(feature1, feature2):
    fig = plt.figure(title="%s vs %s Relation"%(feature1.capitalize(), feature2.capitalize()))
    
    options = {'color': dict(label='Group', orientation='vertical', side='right')}
    
    scat = plt.scatter(x=substance[feature1],
                       y=substance[feature2],
                       color=substance["Grouping Level"],
                       axes_options = options,
                       stroke="black", stroke_width=2.0,
                       default_size=50,
                       default_opacities=[0.7])

    plt.xlabel(feature1.capitalize())
    plt.ylabel(feature2.capitalize())

    plt.show()
```

```python
var1 = ipywidgets.Dropdown(options=["Rate", "YTD Rate", "Numerator", "YTD Numerator",\
                                    "Denominator", "YTD Denominator"],\
                           description='Select for x value: ', style = {'description_width': 'initial'});
var2 = ipywidgets.Dropdown(options=["Rate", "YTD Rate", "Numerator", "YTD Numerator",\
                                    "Denominator", "YTD Denominator"],\
                           description='Select for y value: ', style = {'description_width': 'initial'});
```

```python
ipywidgets.interact(create_scatter, feature1=var1, feature2=var2);
```

Summary:

Here is an interactive dashboard that allows users to make a scatterplot based on any two of the numerical variables in the dataset. To use the dashboard, just simply select the variable from the dropdown list. The first dropdown list is for x-axis, and the second dropdown is for y-axis. The different colors in the scatterplot indicate variables with different grouping level. At the bottom of the dashboard, the first button is used for zoom-in/zoom-out, and you can set it as default by clicking the second button.
